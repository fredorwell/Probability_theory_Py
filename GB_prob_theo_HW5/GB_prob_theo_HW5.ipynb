Задача 3
Проведите тест гипотезы.
Продавец утверждает, что средний вес пачки печенья составляет 200 г.
Из партии извлечена выборка из 10 пачек. Вес каждой пачки составляет:
202, 203, 199, 197, 195, 201, 200, 204, 194, 190.
Известно, что их веса распределены нормально.
Верно ли утверждение продавца, если учитывать, что доверительная вероятность равна 99%? (Провести двусторонний тест.)
Ответ:
Из условия задачи получается что нулевая гипотеза у нас mu=mu0=200, а вторая гипотиза mu!=mu0.
Уровень значимости у нас alpha = 1-0.99 = 0.001
И поскольку нам не известны дисперсия и среднее, то мы воспользуемся критерием Стьюдента.
Нам необходимо посчитать t-критерий и он равен -1.065
Теперь мы можем найти табличное значение для alpha/2=0.01/2= 0.005 и оно равно 3.249
Получается что у нас t1 = 3.249, а t2 = -3.249
Наше значение t наблюдаемое = -1,065 лежит между 3.249 и -3.249
Соответственно мы не отвергаем нулевую гипотезу

```python
n_mean = (202+203+199+197+195+201+200+204+194+190)/10
print(f'Среднее: {n_mean}')
n = numpy.array([202, 203, 199, 197, 195, 201, 200, 204, 194, 190])
n_std = numpy.std(n,ddof=1) 
print(f'Стандартное отклонение: {n_std}')
# Тут пришлось схалтурить и посчитать стандартное отклонение с помощью нампай
tn=(n_mean-200)/((n_std)/10**0.5)
print(f't-критерий: {tn}')
t_table=scipy.stats.t.ppf(0.995, len(n)-1)
print(f't-критерий: {t_table}')
```

Задача 4
Есть ли статистически значимые различия в росте дочерей?
Рост матерей 172, 177, 158, 170, 178,175, 164, 160, 169, 165
Рост взрослых дочерей: 173, 175, 162, 174, 175, 168, 155, 170, 160
Ответ:
Примем alpha = 0.1
pvalue в нашем случае равен 0.5911
соответственно поскольку pvalue = 0.5911 больше чем alpha = 0.1, то мы можем утвердждать что статистически значимых различий нет.

```python
x_1 = numpy.array([172, 177, 158, 170, 178, 175, 164, 160, 169])
x_2 = numpy.array([173, 175, 162, 174, 175, 168, 155, 170, 160])

scipy.stats.ttest_rel(x_1, x_2)
```
